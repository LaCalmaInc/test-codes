```python
import pandas as pd
```

```python
data = {
    'Nombre': ['Ana', 'Luis', 'Carlos', 'Sara'],
    'Edad': [25, 30, 22, 27],
    'Ciudad': ['Madrid', 'Barcelona', 'Valencia', 'Bilbao']
}

df = pd.DataFrame(data)
df.head()
```

```python
df['Edad_en_10_años'] = df['Edad'] + 10
df['Ciudad'] = df['Ciudad'].str.upper()
df['Es_mayor_de_25'] = df['Edad'] > 25
df
```
